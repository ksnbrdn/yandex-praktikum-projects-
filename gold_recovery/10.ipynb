<b>Описание проекта</b>  

Проект, предоставленый компанией «Цифра» (https://www.zyfra.com/ru/), представляет собой реальную задачу Data Science из золотодобывающей отрасли. Компания разрабатывает решения для эффективной работы промышленных предприятий.  
Необходимо подготовить прототип модели машинного обучения для «Цифры». Модель должна предсказать коэффициент восстановления золота из золотосодержащей руды. Модель должна помогать оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками. Имеются данные с параметрами добычи и очистки.

Описание данных  

Данные находятся в трёх файлах:

    gold_recovery_train_new.csv — обучающая выборка;
    gold_recovery_test_new.csv — тестовая выборка;
    gold_recovery_full_new.csv — исходные данные.

Данные индексируются датой и временем получения информации (признак date). Соседние по времени параметры часто похожи.
Некоторые параметры недоступны, потому что замеряются и/или рассчитываются значительно позже. Из-за этого в тестовой выборке отсутствуют некоторые признаки, которые могут быть в обучающей. Также в тестовом наборе нет целевых признаков.
Исходный датасет содержит обучающую и тестовую выборки со всеми признаками.

Технологический процесс:

• Rougher feed — исходное сырье  
• Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant  
	○ Xanthate **— ксантогенат (промотер, или активатор флотации);  
	○ Sulphate — сульфат (на данном производстве сульфид натрия);  
	○ Depressant — депрессант (силикат натрия).  
• Rougher process (англ. «грубый процесс») — флотация  
• Rougher tails — отвальные хвосты  
• Float banks — флотационная установка  
• Cleaner process — очистка  
• Rougher Au — черновой концентрат золота  
• Final Au — финальный концентрат золота  

Параметры этапов:

• air amount — объём воздуха  
• fluid levels — уровень жидкости  
• feed size — размер гранул сырья  
• feed rate — скорость подачи    

Наименование признаков должно быть такое: [этап].[тип_параметра].[название_параметра]  
Пример: rougher.input.feed_ag  

Возможные значения для блока [этап]:  

• rougher — флотация  
• primary_cleaner — первичная очистка  
• secondary_cleaner — вторичная очистка  
• final — финальные характеристики  

Возможные значения для блока [тип_параметра]:  

• input — параметры сырья  
• output — параметры продукта  
• state — параметры, характеризующие текущее состояние этапа  
• calculation — расчётные характеристики

### Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака rougher.output.recovery. Найдём MAE между нашими расчётами и значением признака. 

```python
# Запишем признаки для подсчета в переменные
C = data_train['rougher.output.concentrate_au']
F = data_train['rougher.input.feed_au']
T = data_train['rougher.output.tail_au']
```

```python
# Рассчитаем эффективность обогащения по формуле
data_train['rougher.output.recovery.calculated'] = (C * (F - T)) / (F * (C - T)) * 100
```

```python
# Найдём MAE между нашими расчётами и значением признака
mae = (data_train["rougher.output.recovery.calculated"] - data_train["rougher.output.recovery"]).abs().sum() / len(data_train)
print("MAE", ":", mae)
```

Полученное значение МАЕ достаточно мало. Значит, что эффективность обогащения рассчитана правильно.

### Исследуем суммарную концентрацию всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах. 

```python
# Напишем функцию для рассчета суммарных концентраций на разных этапах очистки
def total_concentration(data):
    # Суммарная концентрация веществ на этапе флотации
    data['rougher.output.concentrate_ag_pb_sol_au'] = data[['rougher.output.concentrate_ag',
                                                            'rougher.output.concentrate_pb',
                                                            'rougher.output.concentrate_sol',
                                                            'rougher.output.concentrate_au']].sum(axis = 1)
    # Суммарная концентрация веществ на этапе первичной очистки
    data['rougher.input.feed.concentrate_ag_pb_sol_au'] = data[['rougher.input.feed_ag',
                                                                    'rougher.input.feed_pb',
                                                                    'rougher.input.feed_sol',
                                                                    'rougher.input.feed_au']].sum(axis = 1)
    # Суммарная концентрация веществ на финальном этапе очистки
    data['final.output.concentrate_ag_pb_sol_au'] = data[['final.output.concentrate_ag',
                                                          'final.output.concentrate_pb',
                                                          'final.output.concentrate_sol',
                                                          'final.output.concentrate_au']].sum(axis = 1)
    return data
```

## Построим модель

### Напишем функцию для вычисления итоговой sMAPE

```python
# Напишем функцию для вычисления sMape
def get_smape(target, predict):
    smape = (1 / len(target)) * np.sum((np.abs(target - predict))/( (np.abs(target) + np.abs(predict)) / 2)) * 100
    return smape
```

```python
smape_scorer = make_scorer(get_smape)
```

```python
# Напишем функцию для вычисления итоговой sMAPE
def get_final_smape(rougher, final):
    final = 0.25 * rougher + 0.75 * final
    return final
```

### Обучим разные модели и оценим их качество кросс-валидацией. Выберем лучшую модель и проверим её на тестовой выборке. 

```python
# Удалим столбцы из data_train, которые не нужны для дальнейшего обучения модели
data_train = data_train.drop(data_train.columns.difference(data_test.columns), axis=1)
```

```python
feature_scaler = StandardScaler()
# Масштабируем численные признаки тренировочной выборки
features_train = feature_scaler.fit_transform(features_train)
# Масштабируем численные признаки тестовой выборки
features_test = feature_scaler.transform(features_test)
```

Обучим модели DecisionTreeRegressor, RandomForestRegressor и LinearRegression и оценили их качество кросс-валидацией. Для каждой модели посчитаем smape и выберем лучшую.

```python
# Обучим модель линейной регрессии
model_lr_rougher = LinearRegression()
smape_rougher = cross_val_score(model_lr_rougher,
                          features_train,
                          target_train_rougher,
                          scoring=smape_scorer,
                          cv=5)
final_smape_rougher = smape_rougher.mean()
model_lr_final = LinearRegression()
smape_final = cross_val_score(model_lr_final,
                        features_train,
                        target_train_final,
                        scoring=smape_scorer,
                        cv=5)
final_smape_final = smape_final.mean()
result_smape = get_final_smape(final_smape_rougher, final_smape_final)
print('smape наилучшей модели:', round((result_smape),3))
```

```python
# Обучим модель дерева решений
best_result_dt = 100
best_depth_rougher_dt = 0
best_depth_final_dt = 0

for depth_rougher in tqdm(range(1,21,1)):
    for depth_final in range(1,21,1):
        model_dt_rougher = DecisionTreeRegressor(max_depth=depth_rougher, random_state=12345)
        smape_rougher = cross_val_score(model_dt_rougher,
                                        features_train,
                                        target_train_rougher,
                                        scoring=smape_scorer,
                                        cv=5)
        final_smape_rougher = smape_rougher.mean()

        model_dt_final = DecisionTreeRegressor(max_depth=depth_final, random_state=12345)
        smape_final = cross_val_score(model_dt_final, 
                                      features_train,
                                      target_train_final,
                                      scoring=smape_scorer,
                                      cv=5)
        final_smape_final = smape_final.mean()
    
        result_smape = get_final_smape(final_smape_rougher, final_smape_final)
        if result_smape < best_result_dt:
            best_result_dt = result_smape
            best_depth_rougher_dt = depth_rougher
            best_depth_final_dt = depth_final
        
print('smape наилучшей модели:', round((best_result_dt),3),
      ',','c лучшим значением для rougher: max_depth =', best_depth_rougher_dt,
      'и','c лучшим значением для final: max_depth =', best_depth_final_dt
     )
```

```python
# Обучим модель случайного леса
best_result_rf = 100
best_est_rf = 0
best_depth_rf = 0
for est in tqdm(range(1,51,10)):
    for depth in range(2,11,2):
        model_rf_rougher = RandomForestRegressor(n_estimators=est, max_depth=depth, random_state=12345)
        smape_rougher = cross_val_score(model_rf_rougher,
                                  features_train,
                                  target_train_rougher,
                                  scoring=smape_scorer,
                                  cv=5)
        final_smape_rougher = smape_rougher.mean()

        model_rf_final = RandomForestRegressor(n_estimators=est, max_depth=depth, random_state=12345)
        smape_final = cross_val_score(model_rf_final,
                                features_train,
                                target_train_final,
                                scoring=smape_scorer,
                                cv=5)
        final_smape_final = smape_final.mean()
        result_smape = get_final_smape(final_smape_rougher, final_smape_final)
        if result_smape < best_result_rf:
            best_result_rf = result_smape
            best_depth_rf = depth
            best_est_rf = est
print('smape наилучшей модели:', round((best_result_rf),3),
      ',','с лучшим значением n_estimators =', best_est_rf,
      ' и лучшим значением max_depth =', best_depth_rf
      )   
```

Лучшая модель - RandomForestRegressor с показателем smape - 8.117, с лучшим значением n_estimators = 41 и лучшим значением max_depth = 4
